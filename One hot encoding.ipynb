```python
#using scikit learn we can use onehotencoder() (OHE)
#using pandas we can use :pandas.get_dummies() 
#is used for data manipulation. 
#It converts categorical data into dummy or indicator variables.

pd.get_dummies(data, drop_first=True)
```

```python
#now just for 6 columns,100+cols are added for onehot encoding.What if we have more cols
#according to studies, it has been proved that we consider top 10 or top 20 of most 
#frequent data occuring, then apply OHE

#from data take 2nd col X2 and apply value_counts: 
#method returns a Series containing the counts of unique values. 
#This means, for any column in a dataframe

data.X2.value_counts()
```

```python
#sort it in descending order and use head(20) so that we get top 20 

data.X2.value_counts().sort_values(ascending= False).head(20)

# here as category has 1658 result, ae etc. 
# apart from these 20, consider all as 0 
```

```python
# MAKING 10 BINARY VARIABLE TO APPLY OHE

#store these top 10 data into a var. tke index

top_10= data.X2.value_counts().sort_values(ascending= False).head(10).index
top_10
```

```python
#use for loop and store in list 

top_10=[x for x in data.X2.value_counts().sort_values(ascending= False).head(10).index]
top_10 #shows top 10
```

```python
#now apply OHE i.e, making all variables in matrix format

#consider for loop that runs through top_10 and get the top_10 result
#now if top_10 is there in X2 col's data: data[x2] then mark 1 else 0 


for label in top_10:
    data[label] = np.where(data['X2'] == label,1,0) #here if my data[x2] is coming under top10 make 1 else 0


data[['X2']+top_10].head(40) #make 2D array where 1st col is X2 and then add all top_10 as rows and cols

# check aq, ag: these are not in top_10 that means all the values will be 0
```
